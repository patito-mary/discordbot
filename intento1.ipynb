```python
import discord
import requests
```

El método __init__(self) es el constructor de la clase. El constructor se utiliza para inicializar la clase. En este caso, el constructor llama al constructor de la clase discord.Client.

El método on_ready() se ejecuta cuando el bot está listo. El método on_ready() se puede utilizar para realizar tareas iniciales, como unirse a canales o enviar mensajes de bienvenida.

El método on_message() se ejecuta cuando el bot recibe un mensaje. El método on_message() se puede utilizar para responder a los mensajes de los usuarios.
